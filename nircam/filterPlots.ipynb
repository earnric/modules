```python
%matplotlib inline
import gc
%config InlineBackend.figure_format = 'retina'
from matplotlib.pylab import *
import matplotlib.pyplot as plt 
rcParams['figure.figsize'] = (12,9)
rcParams['font.size'] = 34
import os
import glob
```

```python
from matplotlib.ticker import FuncFormatter
from matplotlib.ticker import FormatStrFormatter

fmt = '%0.1f'

def log_formatter(x,pos):
    return format(np.log10(x), '.0f')
logfmt  = FuncFormatter(log_formatter)

def div_formatter(x,pos):
    return format(x/1000, '.0f')
divfmt  = FuncFormatter(div_formatter)

# formatter = FormatStrFormatter(fmt)
```

```python
pathToFilterFiles = '/Users/earnric/Google Drive/ASU/Research-Observability/nircam/f*.trans'
allFiles = glob.glob(pathToFilterFiles)
print(allFiles)
```

```python
# Build a python array of numpy arrays. 
# Can't be a numpy array at top level since all dimensions would have to be equal
filters = [np.loadtxt(file,skiprows=1) 
           for file in allFiles]
print(len(filters))
```

```python
pltRect = [0.1, 0.1, .85, 0.85]
ax1     = plt.axes(pltRect)

#ax1.set_xscale('log')
ax1.set_xlim([4e3,3e4])

ax1.set_xlabel('$\lambda$/1000 [$\AA$]', size=24)
ax1.set_ylabel(r'tranmission %', size=26)

ax1.grid(True,which="both",ls=":")
ax1.tick_params('both', length=10, width=2, which='major')
ax1.tick_params('both', length=5, width=1, which='minor')

ax1.plot(filters[0][:,1],filters[0][:,2], 
        filters[1][:,1],filters[1][:,2], 
        filters[2][:,1],filters[2][:,2], 
        filters[3][:,1],filters[3][:,2], 
        filters[4][:,1],filters[4][:,2], 
        filters[5][:,1],filters[5][:,2], 
        filters[6][:,1],filters[6][:,2],
        filters[7][:,1],filters[7][:,2], 
        filters[8][:,1],filters[8][:,2], 
        filters[9][:,1],filters[9][:,2])

ax1.xaxis.set_major_formatter(divfmt)

plt.show()
```
